```python
def hasURL(tweet):
    if len(tweet['entities']['urls']) == 0:
        return False
    else: 
        return True
```

```python
def topURLs(df):
    #create lang dictionary to get a count of the number of times a url is tweeted
    tweet_URLs = {}
    url_to_tweet_map = {}

    for i in range(len(df)):
        if hasURL(df.loc[i]):
            expURL = df.loc[i]['entities']['urls'][0]['expanded_url']
            if expURL.startswith('https://twitter.com/'):
                continue
            else: 
                if expURL in tweet_URLs:
                    tweet_URLs[expURL] += 1
                else:
                    tweet_URLs[expURL] = 1

                if expURL in url_to_tweet_map:
                    url_to_tweet_map[expURL].append(i)
                else:
                    url_to_tweet_map[expURL] = []
                    url_to_tweet_map[expURL].append(i)
        else:
            continue;
            
    return tweet_URLs, url_to_tweet_map
```

```python
#create lang dictionary
tweet_source_URLs = {}
tweet_ext_URLs = {}
countURLS = 0 #have a counter to see how many tweets reference a url link

#set path to access all json files in January folder
data_dir = Path('..') / 'months/feb/02-01to02-04'
data_files = data_dir.glob('*.json')

#Iterate through the json files with the two main for loops below
for file in data_files:
    print(file)
    df = pd.read_json(str(file), lines=True)
    
    for i in range(len(df)):
        if df.loc[i]['lang'] == 'en':
            if len(df.loc[i]['entities']['urls']) == 0:
                continue
            else:
                countURLS += 1

                #Cut up the display url to get the baseurl which should hint towards which source published the link
                dispurl = df.loc[i]['entities']['urls'][0]['display_url']
                urlbase = dispurl.split('/')
                urlsource = urlbase[0]
                expURL = df.loc[i]['entities']['urls'][0]['expanded_url']
                if urlsource in tweet_source_URLs:
                    tweet_source_URLs[urlsource] += 1
                else:
                    tweet_source_URLs[urlsource] = 1

                if expURL.startswith('https://twitter.com/'):
                    continue
                else: 
                    if expURL in tweet_ext_URLs:
                        tweet_ext_URLs[expURL] += 1
                    else:
                        tweet_ext_URLs[expURL] = 1
        else:
            continue;
```

```python
#Turn tweet_source_urls into dataframes then sort by counts
columns = ['url', 'count']
tweet_source_df = pd.DataFrame(tweet_source_URLs.items(), columns=columns)
tweet_source_df = tweet_source_df.sort_values('count', ascending = False)

#add percentage column to dataframe
tweet_source_df['percentage'] = 0.0

#find a total count to base percentage on
totalSum = 0.0
for i in range(len(tweet_source_df)):
    totalSum += tweet_source_df.loc[i]['count']

#calculate percentage for each column and language name
for i in range(len(tweet_source_df)):
    tweet_source_df.at[i, 'percentage'] = tweet_source_df.loc[i]['count'] / totalSum * 100.0

#sort the dataframe by percentage column in ascending order
tweet_source_df = tweet_source_df.sort_values(by=['percentage'], ascending=False)

#Grouping less than 1 percent languages into other dataframe
others_df = tweet_source_df.loc[tweet_source_df["percentage"] < 1.0]
other_count = sum(others_df['count'].tolist()) #Get a total for all languages under 1 percent

#get indices for all languages under 1 percent to drop from result_df
dropList = others_df.index.tolist()
result_df = tweet_source_df.drop(dropList)

#add other to result_df
otherData = {'url': 'other', 'count': other_count, 'percentage': other_count / totalSum * 100.0}
result_df = result_df.append(otherData, ignore_index=True)
result_df = result_df.sort_values('count', ascending=False)
result_df
```
