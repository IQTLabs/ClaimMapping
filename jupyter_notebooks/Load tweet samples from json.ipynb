```python
import pandas as pd
from pathlib import Path
pd.set_option('display.max_colwidth', 800)
data_dir = Path('../..') / 'data_samples/tweet_examples_echen.json'
f = str(data_dir)
df = pd.read_json(f)
```

```python
# How to access very nested subfields
df.loc['user'][1]['profile_image_url']
```

```python
# df.to_excel('sample_all.xlsx')
```

```python
df.index
```

```python
row_to_keep = ['created_at','id','full_text','in_reply_to_user_id','geo','place','retweet_count','lang']
```

```python
col_to_keep = [1,3,4]
```

```python
# View the columns we want to save for a sample: 
df.loc[row_to_keep][col_to_keep]
```

```python
df_save = df.loc[row_to_keep][col_to_keep]
```

```python
df_user = pd.DataFrame.from_dict(df.loc['user'][1])
```

```python
df_user
```

```python
for i in range(3,5): # We are skipping #2 b/c it's not in the text dataframe
    df_user = df_user.append(pd.DataFrame.from_dict(df.loc['user'][i]))
```

```python
df_user # This outputs a duplicate row for each user who used a url
```

```python
df_user = df_user.transpose() #Reorient to original
```

```python
df_user.columns = [1,3,'4_dup',4]
```

```python
df_user = df_user.drop('4_dup',axis = 1)
```

```python
df_user.index
```

```python
user_row_to_keep = ['id','name','screen_name','created_at','location','description','profile_image_url','followers_count','friends_count']
```

```python
df_user.loc[user_row_to_keep]
```

```python
final_save = pd.concat([df_save,df_user.loc[user_row_to_keep]],axis=0)
```

```python
final_save #Everything (2nd) id down is user info
```

```python
final_save.index
```

```python
# Rename some rows
final_save.index = ['created_at', 'id', 'full_text', 'in_reply_to_user_id', 'geo', 'place',
       'retweet_count', 'lang', 'user_id', 'user_name', 'screen_name', 'user_created_at','location',
       'description', 'profile_image_url', 'followers_count', 'friends_count']
```
